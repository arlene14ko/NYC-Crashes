### Consolidating the street names under 1 new column `street_name`

```python
def clean_street(i): 
    street = str(i)
    if street == "nan":
        return i
    else:
        name = street.upper().split("    ")
        for x in range(len(name)):
            if re.search('[a-zA-Z]+$', name[x]):
                return name[x].strip()
        
        
def clean(i):
    street = str(i)
    if street == "nan": 
        return i 
    else: 
        street = str(i)
        return street.upper().strip()
```

### Consolidating zipcode with Location using geopy

```python
from geopy.geocoders import Nominatim
```

```python
geolocator = Nominatim(user_agent="NYC-Crashes-draft-notebook")
location = geolocator.reverse("40.536728, -74.193344")
location.raw
```

```python
def get_zip(x):
    zipcode = str(x)
    if zipcode == "nan":
        return x
    else:
        li = zipcode.split(" ")
        for i in range(len(li)):
            li[i] = li[i].strip("(,)")
            
        x = float(li[0])
        y = float(li[1])
    
        location = geolocator.reverse(f"{x},{y}")
        return location.raw['address']['postcode']
```

```python
get_zip("(40.5368705, -74.19332961232394)")
```
